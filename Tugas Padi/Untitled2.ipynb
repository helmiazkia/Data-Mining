```python
import pandas as pd

# Membaca file Excel yang diunggah
file_path = '/produksiPadi.xlsx'
data = pd.read_excel(file_path)

# Menampilkan beberapa baris pertama data untuk melihat strukturnya
data.head()
```

```python
# Membersihkan data: Memilih hanya kolom 'Tahun' dan 'Produksi Padi(Ton)'
data = data[['Tahun', 'Produksi Padi(Ton)']].dropna()

# Mengganti nama kolom untuk kemudahan analisis
data.columns = ['Tahun', 'Produksi']

# Menampilkan data yang telah dibersihkan
data.head()
```

```python
# Membuat data lagging: Produksi tahun sebelumnya (lagging 1 tahun)
data['Produksi_Lag'] = data['Produksi'].shift(1)

# Menghapus baris dengan nilai NaN (baris pertama akan kosong setelah lagging)
data_lagged = data.dropna()

# Menampilkan data hasil lagging
data_lagged.head()
```

```python
# Import library yang diperlukan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import numpy as np

# Membaca file Excel
file_path = 'produksiPadi.xlsx'  # Ganti dengan path file Anda
data = pd.read_excel(file_path)

# Membersihkan data
data = data[['Tahun', 'Produksi Padi(Ton)']].dropna()
data.columns = ['Tahun', 'Produksi']

# Membuat data lagging: Produksi tahun sebelumnya
data['Produksi_Lag'] = data['Produksi'].shift(1)
data_lagged = data.dropna()

# Memisahkan fitur (X) dan target (y)
X = data_lagged[['Produksi_Lag']]
y = data_lagged['Produksi']

# Membagi data menjadi training (70%) dan testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=False)

# Melatih model regresi linear
model = LinearRegression()
model.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred = model.predict(X_test)

# Evaluasi model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print("Hasil Evaluasi Model:")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2%}")

# Prediksi untuk tahun 2023 dan 2024
tahun_2022_produksi = data_lagged.iloc[-1]['Produksi']  # Data produksi terakhir (2022)
produksi_2023 = model.predict([[tahun_2022_produksi]])[0]  # Prediksi untuk 2023
produksi_2024 = model.predict([[produksi_2023]])[0]  # Prediksi untuk 2024

print("\nPrediksi Produksi Padi:")
print(f"Tahun 2023: {produksi_2023:.2f} ton")
print(f"Tahun 2024: {produksi_2024:.2f} ton")
```
